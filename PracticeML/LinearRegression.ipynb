# Linear Regression

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
import pandas as pd
df = pd.read_csv('50_Startups.csv')
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

#### Establish relationship between Profit vs RND

```python
# visualising profit and RND
import matplotlib.pyplot as plt
plt.scatter(df['RND'],df['PROFIT'])
plt.xlabel('RND')
plt.ylabel('PROFIT')
plt.title('PROFIT vs RND')
plt.grid()
plt.show()
```

#### Separate Dependent and Independent Features

```python
X = df[['RND']]
Y = df[['PROFIT']]
```

```python
X.head()
```

```python
Y.head()
```

### Build a Linear Regression model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,Y)
```

```python
model.coef_
```

```python
model.intercept_
```

### Equation of RND vs Profit

PROFIT = 49032.89914125 + 0.85429137 *RND

```python
ypred = model.predict(X)
```

```python
ypred
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['RND'],df['PROFIT'])
plt.plot(X,ypred,c='red')
plt.xlabel('RND')
plt.ylabel('PROFIT')
plt.title('PROFIT vs RND')
plt.grid()
plt.show()
```

## Evaluate Model
1.  Mean Squared Error
2.  Root mean squared error
3.  Mean absolute error
4.  R2 score

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
mse = mean_squared_error(Y,ypred)
rmse = mse**(1/2)
print(f'Mean Squared Error : {mse:.2f}')
print(f'Root Mean Squared Error : {rmse:.2f}')
```

```python
mae = mean_absolute_error(Y,ypred)
print(f'Mean Absolute Error : {mae :.2f}')
```

```python
r2 = r2_score(Y,ypred)
print(f'R2 score : {r2:.4f}')
```
