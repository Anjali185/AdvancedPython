# Creating custom class person

```python
class Person:

    # Constructor function - define attributes of function
    def __init__(self,name,age,gender,occupation):
        self.name = name
        self.age = age
        self.gender = gender
        self.occupation = occupation
    
    # User Defined Function
    def talk(self):
        print(f'Hi There my name is {self.name}')
        print(f'My age is {self.age}')
        print(f'I am {self.gender}')
        print(f'I work as {self.occupation}')
```

```python
p1 = Person(name = 'Anjali',age = 34 , gender = 'Female' , occupation = 'Data Scientist')
```

# Create a class Bank account



```python
class account:

    # Add init constructor
    def __init__(self,ac_no,name,bal=0):
        self.ac_no = ac_no
        self.name = name
        self.bal = bal
    
    # Function for depositing money
    def depositMoney(self,amt):
        self.bal = self.bal + amt
        print(f'Amount {amt} INR is deposited into your account number : {self.ac_no}')
        print(f'Current Balance is {self.bal} INR')
    
    # Function to withdraw money
    def withdrawMoney(self,amt):
        if self.bal >= amt:
            self.bal = self.bal - amt
            print(f'Amount {amt} INR is withdrawn from your account number : {self.ac_no}')
            print(f'Current Balance is {self.bal} INR')
        else:
            print(f'Insufficient Balance in account {self.ac_no}. Current Balance in your account is {self.bal} INR')

    # Function to transfer money
    def transferMoney(self,amt,ac2):
        if self.bal >= amt:
            self.bal = self.bal-amt
            ac2.bal = ac2.bal +amt
            print(f'Amount {amt} INR is transferred from your account number : {self.ac_no} to account : {ac2.ac_no}')
            print(f'Balance {self.bal} INR')
        else:
            print(f'Insufficient Balance in account {self.ac_no}. Current Balance in your account is {self.bal} INR')
    
        
```

```python
a1 = account(ac_no=123,name='Jiten Shah',bal = 6000)
```

```python
a2 = account(ac_no=456,name='Lalit Sharma')
```

# Assignment for Students



```python
class triangle:

    # Add init constructor
    def __init__(self,a,b,c):
        self.a = a
        self.b = b
        self.c = c

    # To check does Sides form a triangle
    def CheckIfTriangle(self):
        if (((self.a + self.b) > self.c) and ((self.b + self.c) > self.a) and ((self.c + self.a) > self.b)) :
            tr = True
            print(f'This is a Triangle')
        else:
            tr = False
            print(f'This is NOT a triagnle')
        return tr
    
    # Perimeter of Triangle
    def PerimeterTriangle(self):
        if self.CheckIfTriangle() == True:
            p = self.a + self.b + self.c
            print(f'Perimeter of Triangle: {p}')
        else:
            p=None

    # Check Area of Triangle with herons formula
    def AreaTriangleHerons(self):
        if self.CheckIfTriangle() == True:
            semiPeri = (self.a+self.b+self.c)/2
            print(f'Semi Perimeter is :{semiPeri}')
            areaT = (semiPeri*(semiPeri-self.a)*(semiPeri-self.b)*(semiPeri-self.c))**0.5
            print(f'Ares of Triangle with Herons Formula is : {areaT}')
        else:
            areaT = None
        return areaT
```
